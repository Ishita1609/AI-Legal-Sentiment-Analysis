```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
import pandas as pd

data = {
    "id": [1, 2, 3, 4, 5],
    "text": [
        "The agreement benefits both parties involved.",
        "The defendant is ordered to pay damages within 30 days.",
        "This contract outlines responsibilities in detail.",
        "Failure to comply will result in heavy penalties.",
        "The settlement was mutually agreed upon and accepted."
    ]
}

df = pd.DataFrame(data)
df.to_csv("legal_docs.csv", index=False)
df.head()
```

```python
results = []
for text in df["text"]:
    result = sentiment_pipeline(text)[0]
    results.append(result["label"])  # Example output: '4 stars'
```

```python
df["stars"] = results
```

```python
def map_sentiment(star_label):
    stars = int(star_label[0])
    if stars <= 2:
        return "NEGATIVE"
    elif stars == 3:
        return "NEUTRAL"
    else:
        return "POSITIVE"

df["sentiment"] = df["stars"].apply(map_sentiment)
df.head()
```

```python
df.to_csv("labeled_outputs.csv", index=False)
```
